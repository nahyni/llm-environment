```python
desired_columns = ['자치구', '행정동', '주거인구', '인구밀도', '유동인구', '노인인구수','등록차량 개수','지하철 승객수',
                   '따릉이반납자수','따릉이 개수','녹지 개수','건물 개수','대기환경시설개수','ALL_CO2']

# 컬럼 순서 재정렬
df_merged = df_merged[desired_columns]
df = df_merged
```

```python
from sklearn.preprocessing import MinMaxScaler

# 제외할 컬럼을 리스트로 정의
exclude_columns = ['자치구', '행정동', 'ALL_CO2']

# 숫자형 컬럼만 선택 (특정 컬럼 제외)
X_columns = df.columns[~df.columns.isin(exclude_columns)]

# Min-Max 스케일러 초기화
scaler = MinMaxScaler()

# 숫자형 컬럼에 대해 스케일링 적용
df[X_columns] = scaler.fit_transform(df[X_columns])

# 결과 출력
df
```

```python
df.columns
```

```python
from sklearn.ensemble import RandomForestRegressor
import pandas as pd

# 독립 변수 (X)와 종속 변수 (y) 설정
X_pop = df[['주거인구', '인구밀도', '유동인구', '노인인구수']]
X_traffic = df[['등록차량 개수', '지하철 승객수', '따릉이반납자수', '따릉이 개수']]
X_infra = df[['녹지 개수', '건물 개수','대기환경시설개수']]
y = df['ALL_CO2']
```

### 인구지수

```python
# 랜덤 포레스트 모델 학습
rf = RandomForestRegressor(random_state=42)
rf.fit(X_pop, y)

# feature importance 추출
feature_importance = rf.feature_importances_

# 각 변수에 가중치를 적용하여 인구지수 계산
df['인구지수'] = (X_pop['주거인구'] * feature_importance[0] +
                 X_pop['인구밀도'] * feature_importance[1] +
                 X_pop['유동인구'] * feature_importance[2] +
                 X_pop['노인인구수'] * feature_importance[3])

# 결과 확인
# 가중치 값을 출력하는 코드
print(f'주거인구 가중치: {feature_importance[0]}\n'
      f'인구밀도 가중치: {feature_importance[1]}\n'
      f'유동인구 가중치: {feature_importance[2]}\n'
      f'노인인구수 가중치: {feature_importance[3]}')

print(df[['자치구', '행정동', '인구지수']])
```

### 교통 지수

```python
# 랜덤 포레스트 모델 학습
rf = RandomForestRegressor(random_state=42)
rf.fit(X_traffic, y)

# feature importance 추출
feature_importance = rf.feature_importances_

# 각 변수에 가중치를 적용하여 인구지수 계산
df['교통지수'] = (X_traffic['등록차량 개수'] * feature_importance[0] +
                 X_traffic['지하철 승객수'] * feature_importance[1] +
                 X_traffic['따릉이반납자수'] * feature_importance[2] +
                 X_traffic['따릉이 개수'] * feature_importance[3] )

# 결과 확인
# 가중치 값을 출력하는 코드
print(f'등록차량 가중치: {feature_importance[0]}\n'
      f'지하철 승객수 가중치: {feature_importance[1]}\n'
      f'따릉이 반납자 수 가중치: {feature_importance[2]}\n'
      f'따릉이 개수 가중치: {feature_importance[3]}\n')

print(df[['자치구', '행정동', '교통지수']])
```

### 인프라 지수

```python
# 랜덤 포레스트 모델 학습
rf = RandomForestRegressor(random_state=42)
rf.fit(X_infra, y)

# feature importance 추출
feature_importance = rf.feature_importances_

# 각 변수에 가중치를 적용하여 인구지수 계산
df['인프라지수'] = (X_infra['녹지 개수'] * feature_importance[0] +
                 X_infra['건물 개수'] * feature_importance[1] +
                  X_infra['대기환경시설개수'] * feature_importance[2] )

# 결과 확인
# 가중치 값을 출력하는 코드
print(f'녹지 가중치: {feature_importance[0]}\n'
      f'건물 가중치: {feature_importance[1]}\n'
      f'대기환경시설 가중치: {feature_importance[2]}')

print(df[['자치구', '행정동', '인프라지수']])
```

## 클러스터링

```python
# 필요한 라이브러리 불러오기
import pandas as pd
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# 데이터 준비 (필요한 컬럼만 선택)
X = df[['인구지수', '교통지수', '인프라지수']]

# 스케일링 (클러스터링 전 정규화)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# K-Means 클러스터링
kmeans = KMeans(n_clusters=3, random_state=42)
df['kmeans_cluster'] = kmeans.fit_predict(X_scaled)

# 계층적 클러스터링
hierarchical = AgglomerativeClustering(n_clusters=3)
df['hierarchical_cluster'] = hierarchical.fit_predict(X_scaled)

# DBSCAN 클러스터링
dbscan = DBSCAN(eps=0.5, min_samples=5)
df['dbscan_cluster'] = dbscan.fit_predict(X_scaled)

# 각 클러스터에 속한 행정동 출력
print("K-Means 클러스터 결과")
print(df[['행정동', 'kmeans_cluster']].groupby('kmeans_cluster').agg(list))

print("\n계층적 클러스터링 결과")
print(df[['행정동', 'hierarchical_cluster']].groupby('hierarchical_cluster').agg(list))

print("\nDBSCAN 클러스터링 결과")
print(df[['행정동', 'dbscan_cluster']].groupby('dbscan_cluster').agg(list))

# 시각화 (필요한 경우)
plt.figure(figsize=(10, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=df['kmeans_cluster'], cmap='viridis')
plt.title("K-Means 클러스터링 결과")
plt.xlabel('인구지수')
plt.ylabel('교통지수')
plt.show()
```

```python
kmeans_result = df[['행정동', 'kmeans_cluster']].groupby('kmeans_cluster').agg(list)
hierarchical_result = df[['행정동', 'hierarchical_cluster']].groupby('hierarchical_cluster').agg(list)
dbscan_result = df[['행정동', 'dbscan_cluster']].groupby('dbscan_cluster').agg(list)
```

```python
import geopandas as gpd
import pandas as pd
from esda import Moran_Local
from libpysal.weights import KNN
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 데이터를 GeoDataFrame으로 로드
# df에는 ['지역구', '인구지수', '교통지수', '인프라지수', 'geometry'] 컬럼이 있어야 함
# geometry는 Polygon 형태의 지리 정보
gdf = gpd.GeoDataFrame(df, geometry='geometry')

# LISA를 위해 사용하는 지표 합산 (인구지수, 교통지수, 인프라지수)
gdf['composite_index'] = gdf[['인구지수', '교통지수', '인프라지수']].mean(axis=1)

# 공간 가중치 행렬을 생성 (k=5 이웃 가중치)
w = KNN.from_dataframe(gdf, k=5)
w.transform = 'r'

# LISA 계산
lisa = Moran_Local(gdf['composite_index'], w)

# 핵심 지역 선정 (High-High 패턴)
gdf['LISA_cluster'] = lisa.q
gdf['significance'] = lisa.p_sim

# 핵심 지역은 High-High 클러스터로, 유의미한 지역만 필터링
core_areas = gdf[(gdf['LISA_cluster'] == 1) & (gdf['significance'] < 0.05)]

# 핵심 지역 3개 선정
core_areas_sorted = core_areas.sort_values('composite_index', ascending=False).head(3)

# 결과 시각화
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
gdf.plot(column='LISA_cluster', cmap='OrRd', linewidth=0.8, ax=ax, edgecolor='0.8')
core_areas_sorted.plot(ax=ax, color='blue', marker='o', markersize=50)
plt.show()

# 핵심 지역 출력
core_areas_sorted[['지역구', 'composite_index']]
```
